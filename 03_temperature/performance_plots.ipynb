```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

points = [10000, 25000, 50000]
processes = [1, 2, 4, 8, 16, 24]

li1 =  [5.154290,
2.915674,
1.586596,
0.933572,
0.579254,
0.510376
]

li2 =  [82.774575,
42.890259,
23.449562,
13.384796,
7.193123,
5.808554
]

li3 =  [369.724277,
167.720056,
90.058679,
52.230787,
25.018365,
20.432366
]
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(processes, li1, '-ob')
plt.legend([10000])
plt.xlabel('Количество потоков')
plt.ylabel('Время, с')
plt.title('Время на выполнение\nв зависимости от количества потоков')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(processes, li2, '-og')
plt.legend([25000])
plt.xlabel('Количество потоков')
plt.ylabel('Время, с')
plt.title('Время на выполнение\nв зависимости от количества потоков')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(processes, li3, '-or')
plt.legend([50000])
plt.xlabel('Количество потоков')
plt.ylabel('Время, с')
plt.title('Время на выполнение\nв зависимости от количества потоков')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(np.log2(processes), np.log2(li1), '-ob')
plt.plot(np.log2(processes), np.log2(li2), '-og')
plt.plot(np.log2(processes), np.log2(li3), '-or')
plt.legend([10000, 25000, 50000])
plt.xlabel('Количество потоков, log2(n)')
plt.ylabel('Время, log2(sec)')
plt.title('Время на выполнение\nв зависимости от количества потоков\n(loglog)')
plt.show()
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(np.log2(processes), np.log2( np.array(li1)/max(li1) ), '-ob')
plt.plot(np.log2(processes), np.log2( np.array(li2)/max(li2) ), '-og')
plt.plot(np.log2(processes), np.log2( np.array(li3)/max(li3) ), '-or')
plt.legend([10000, 25000, 50000])
plt.xlabel('Количество потоков, log2(n)')
plt.ylabel('Время, log2(sec)')
plt.title('Нормализованное время на выполнение\nв зависимости от количества потоков\n(loglog)')
plt.show()
```
